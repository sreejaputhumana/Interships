We have to perform exploratory data analysis on superstore data and we have to find ways to make profit.

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#read data set
store_data = pd.read_csv('C:/Users/sreej/Documents/Python Scripts/SampleSuperstore.csv')
```

```python
#read size of the file
store_data.shape
```

```python
#check for missing values
store_data.isnull().sum()
```

There is no missing values.
From the primary analysis of the data we found country,postalcode,religion are not important factors to make preditions.
Create a copy of data set and remove the above mentioned factors from the data set.


```python
store_prep = store_data.copy()
store_prep = store_prep.drop(['Country','Postal Code','Region'], axis=1)
```

```python
store_prep['Sub-Category'].value_counts().plot.bar(color='b')
plt.xlabel('Sub Category')
plt.title('Sub Category Count')
```

From the above graph,binders and papers are having the highest number and copiers and machines having the lowest count. Then we can check for the profit of each subcategories.

```python
#subcategories and profit
store_prep.groupby('Sub-Category')['Profit'].agg(['sum']).plot.bar(color='b')
plt.xlabel('Sub Category')
plt.ylabel('Profit')
plt.title('Sub Category Vs Profit')
```

Even though copiers have less number of count it has the highest profit rate compared to all other subcategories.
We can see that sales of tables in the store is in loss.


```python
#subcategories and profit
store_prep.groupby('Sub-Category')['Sales'].agg(['sum']).plot.bar(color='b')
plt.xlabel('Sub Category')
plt.ylabel('Sales')
plt.title('Sub Category Vs Sales')
```

When it comes to sales chairs and phones are the items that users are interested to buy from the store.

```python
#segment wise count of products 
store_prep['Segment'].value_counts()
```

The sales are highest in consumer segment.

```python
#profit sale relation state wise

ps_ratio = store_prep.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=True)
ps_ratio.plot.bar(color=['red','blue'],figsize=(15,9))

plt.title("Profit Vs Sale across state")
plt.xlabel("State")
plt.ylabel("Profit & Sale")
```

The stores in California and New York has the highest profit and sale. Even though the sales in both the states are same, profit in California is very high compared to New York.

```python
#discount rate and profit
store_prep.groupby('Discount')['Profit'].agg(['sum']).plot(color='b')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.title('Discount Vs Profit')
```

The items with lowest discount provides a highest profit to the stores.

# Analysis

1) Products like copiers, phones and accessories plays massive roles in company profit. Even though furniture items like tables and chairs are sold in high number, when it comes to profit these are not profitable for the company.

2) Stores in Texas, Pennsylvania, Florida, Illinios, Ohio, North Carolina, Arizona, Tennessee are running in loss.

3) From all these facts and figures we can conclude that the store items that has the highest discount rates have the lowest profit. Companies profit can be inceased by reducing the discount rates of those respective items.

4) Also the profit margin can be increased by increasing the production of products like copiers which is the highest profitable item.
